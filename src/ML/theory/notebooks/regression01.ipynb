```julia
using LinearAlgebra, CairoMakie, NAJ
```

```julia
xs = 1:0.5:10
f(x, p) = p[1]  + p[2] * x^1 +p[3] * x^2 + p[4] * x^3
p0 = [-9.0, 11.8, -3.0, 0.2]
f(x) = f(x, p0)
ys = f.(xs) .+ 1.0*(rand(length(xs)) .- 0.5);
Φ = [x^k for x in xs, k in 0:3]
θ = inv(Φ'*Φ) * Φ' * ys
f1(x) = f(x, θ)
```

```julia
fig = Figure()
ax = Axis(fig[1,1])
scatter!(ax, xs, ys, markersize=10)
lines!(ax, xs, f1.(xs), color = "red")
# save("polyreg.png", fig)
fig
```
