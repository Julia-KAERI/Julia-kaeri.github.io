```julia
fs(x) = 0.04*(x-3)*(x-2)*(x-1.5)*(x+1.5)*(x+2)*(x+3)*exp(-x^2/7)
t = collect(-5:0.1:5)

fig1=Figure(size=(600, 400))

ax1 = Axis(fig1[1, 1], limits=(-5, 5, -5, 10) )

text!(ax1, "text", position=(0,0),     space = :data)
lines!(ax1, t, fs.(t), label="S")


scatter!(ax1, [0.0,],  [fs(0.0), ], color="red")


display(fig1)
# typeof(fig1)
```
